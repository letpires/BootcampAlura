```
import pandas as pd
```

Importação das bibliotecas usadas

Importando os dados diretamente do arquivo csv salvo no computador

Usar o encoding"ISO-8859-1" para poder ler arquivos brasileiros

**skiprows** pula algumas linhas do início do arquivo e **skipfooter** pula as últimas linhas que não são importantes pro dataframe

A separação pra milhar utilizada é o ponto(.) e a separação de casas decimais utilizada é a vírgula (,)



```
dados=pd.read_csv("/content/A221457189_28_143_208.csv", encoding="ISO-8859-1",
                  skiprows = 3, sep=";", skipfooter=12,
                  thousands=".", decimal=",")


dados.head() ## Mostra os 5 primeiros itens da tabela
```

```
dados.tail()  ## Mostra os 5 últimos itens da tabela
```

Para fazer uma visualização dos dados em forma de gráfico é só usar o plot disponível no Pandas. 

```
dados.plot(x="Unidade da Federação", y="2008/Ago")
```

```
dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(10,7))
```

Pra utilizar algumas funções extras, precisamos importar uma biblioteca gráfica, que vai permitir, por exemplo, colocar título no gráfico.

```
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Valor por unidade da federação")
plt.show()
```
